```python
import tensorflow as tf 
import pandas as pd 
import matplotlib.pyplot as plt
import numpy  as np
import cv2
import glob
import os
import pickle
```

```python
training = pickle.load(open("training3", "rb"))
testing = pickle.load(open("testing3", "rb"))
val = pickle.load(open("valing3", "rb"))
training[0][0].shape
```

```python
x_train = []
y_train = []

x_test = []
y_test = []

x_val = []
y_val = []

for feature, label in training : 
    x_train.append(feature)
    y_train.append(label)
    
    
for feature, label in testing : 
    x_test.append(feature)
    y_test.append(label)    
    

for feature, label in val : 
    x_val.append(feature)
    y_val.append(label)    

x_train = np.array(x_train)/255
y_train = np.array(y_train)

x_test = np.array(x_test)/255
y_test = np.array(y_test)

x_val = np.array(x_val)/255
y_val = np.array(y_val)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
```

```python
model = tf.keras.applications.DenseNet121(weights=None, include_top = False, input_shape=(100, 100, 3))
```

```python
predictions = tf.keras.layers.Dense(14, activation='sigmoid',)(model.output)
```

```python
model = tf.keras.Model(inputs=model.input, outputs=predictions)  
```

```python
model.summary()
```

```python
chex_weights_path = 'brucechou1983_CheXNet_Keras_0.3.0_weights.h5'
model.load_weights(chex_weights_path)
model.summary()
```

```python
model.trainable=False
model.training = False
```

```python
last_layer = model.layers[-2]
last_output = last_layer.output
last_output
```

```python
x = tf.keras.layers.GlobalAveragePooling2D()(last_output)
x
```

```python
x = tf.keras.layers.Dense(1, activation='sigmoid')(x)
```

```python
final = tf.keras.Model( model.input, x)
final.summary()
```

```python
final.compile(loss = 'binary_crossentropy',
              optimizer = 'adam',
              metrics = ['binary_accuracy'])
from tensorflow.keras import callbacks  
early_stopping = callbacks.EarlyStopping(
    monitor='val_loss',
    patience=2,
    min_delta=1e-7,
    restore_best_weights=True,
)

plateau = callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor = 0.2,                                     
    patience = 2,                                    
    min_delt = 1e-7,                                
    cooldown = 0,                               
    verbose = 1
) 
history = final.fit(
         x_train,
         y_train,
         epochs= 10,   
    validation_data=(x_test,y_test),
     batch_size = 20 , 
#     steps_per_epoch=100 , 
    callbacks=[early_stopping, plateau]
#     verbose = 2
     
)
```

```python
final.evaluate(x_val, y_val)
```

```python
final.evaluate(x_test, y_test)
```

```python
index = 10
image = x_val[index].reshape(-1,100,100, 3)
re = np.round(final.predict(image))
act = y_val[index]
print(f"the predicatin is {re}")
print(f"the act is {act}")
      
```
